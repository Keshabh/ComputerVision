```python
#lets resize it using pillow
from PIL import Image
img = Image.open('/content/dog')
resized_img = img.resize((200,200))
resized_img.save('dog_image_resized.jpg')
```

```python
# lets read this resized image now
fetched_img = mimage.imread('/content/dog_image_resized.jpg')
plt.imshow(fetched_img)
plt.show()
```

```python
#lets use opencv to perform grayscaling on the image.
img = cv2.imread('/content/dog_image_resized.jpg')
img.shape
```

```python
grayscaleImg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
grayscaleImg.shape
```

```python
#lets visualize this now
plt.imshow(grayscaleImg)
plt.show()
```

```python
#cv2 also has imshow method to display the ndarray to image, but colab does not allow it for some reasons
#instead colab offers that method inside
from google.colab.patches import cv2_imshow
cv2_imshow(grayscaleImg)
```

```python
#lets save the grayscale image
cv2.imwrite('grayscale_image.jpg', grayscaleImg)
```
